```python
#import relevant libraries
import pandas as pd #spreedsheet
import requests #get request
from bs4 import BeautifulSoup as soup #html parser
import urllib.request as  urequest #get request
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
```

```python
request= urequest.Request(url)
```

```python
#open the url
response= urequest.urlopen(request)
```

```python
List_of_largest_companies_by_revenue= response.read()
print(List_of_largest_companies_by_revenue)
```

```python
page= soup(List_of_largest_companies_by_revenue, 'html.parser')
print(page)
```

```python
data=page.find_all('table')
#display num of table
print(f'Number of wikitable tables found:{len(data)}')

#cover the furst table to a DataFrame
if data:
    table=data[0] #Access the first wikitable
    df_list= pd.read_html(str(table))
```

```python
# Access the first DataFrame (usually the one you want)
List_of_largest_companies_by_revenue = df_list[0]
```

```python
List_of_largest_companies_by_revenue
```

```python
col_list = []
for col in List_of_largest_companies_by_revenue.columns:
  col_list.append(col[0])
List_of_largest_companies_by_revenue.columns
```

```python
col_list
```

```python
List_of_largest_companies_by_revenue.columns = col_list
```

```python
#changing the object to int
List_of_largest_companies_by_revenue['Revenue'] = List_of_largest_companies_by_revenue['Revenue'].replace({'\$': '', ',': ''}, regex=True).astype(int)
List_of_largest_companies_by_revenue['Profit'] = List_of_largest_companies_by_revenue['Profit'].replace({'\$': '', ',': ''}, regex=True).astype(int)

print(List_of_largest_companies_by_revenue['Revenue'].head())
```

```python
# Remove the last two columns
List_of_largest_companies_by_revenue = List_of_largest_companies_by_revenue.iloc[:, :-2]
print(List_of_largest_companies_by_revenue.head())
```

**Industry with highest revenue**


```python


# Industry with the highest revenue
industry_revenue = List_of_largest_companies_by_revenue.groupby('Industry')['Revenue'].sum()
# Use idxmax instead of idmax to find the index of the maximum revenue
highest_revenue_industry = industry_revenue.idxmax()
print(f"The industry with the highest revenue is: {highest_revenue_industry}")

# Sorting the industries by total revenue in descending order
industry_revenue = industry_revenue.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_revenue.index, y=industry_revenue.values)
plt.xlabel("Industry")
plt.ylabel("Total Revenue (USD millions)")
plt.title("Total Revenue by Industry")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

**Industry with highest profit**

```python
# Industry with the highest profit
industry_profit = List_of_largest_companies_by_revenue.groupby('Industry')['Profit'].sum()
highest_profit_industry = industry_profit.idxmax()
print(f"The industry with the highest profit is: {highest_profit_industry}")

# Sorting the industries by total profit in descending order
industry_profit = industry_profit.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_profit.index, y=industry_profit.values)
plt.xlabel("Industry")
plt.ylabel("Total Profit (USD millions)")
plt.title("Total Profit by Industry")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

**Total Employees by Industry**

```python
industry_employees = List_of_largest_companies_by_revenue.groupby('Industry')['Employees'].sum()
highest_employees_industry = industry_employees.idxmax()
print(f"The industry with the highest employees is: {highest_employees_industry}")

# Sorting the industries by total profit in descending order
industry_employees = industry_employees.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_employees.index, y=industry_employees.values)
plt.xlabel("Industry")
plt.ylabel("Total Employees")
plt.title("Total Employees by Industry")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

**Total Profit by Ram**

```python
ram_profit = List_of_largest_companies_by_revenue.groupby('Ram')['Profit'].sum()
highest_profit_ram = ram_profit.idxmax()
print(f"The ram with the highest profit is: {highest_profit_ram}")

# Sorting the RAM by total profit in descending order
ram_profit = ram_profit.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=ram_profit.index, y=ram_profit.values)
plt.xlabel("Ram")
plt.ylabel("Total Profit (USD millions)")
plt.title("Total Profit by Ram")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

**Does the number of employees influence profit?**


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Employees', y='Profit', data=List_of_largest_companies_by_revenue)
plt.xlabel('Number of Employees')
plt.ylabel('Profit')
plt.title('Relationship between Number of Employees and Profit')
plt.grid(True)
plt.show()


# Calculate the correlation coefficient
correlation = List_of_largest_companies_by_revenue['Employees'].corr(List_of_largest_companies_by_revenue['Profit'])
print(f"Correlation coefficient between Employees and Profit: {correlation}")
```

A correlation coefficient of
−0.0396 suggests a very weak negative linear relationship between the number of employees and profit. This means that as the number of employees increases, the profit tends to decrease very slightly, but the relationship is so weak that it might not be meaningful.
